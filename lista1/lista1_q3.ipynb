```python
# Exercí­cio 3(f)

import numpy as np
from scipy.optimize import minimize
import matplotlib.pyplot as plt
from numpy import linalg

np.random.seed(1)

beta = np.array([-1.5, 2.0])
input_range = np.linspace(-1, 1, 100)
X = np.vstack([np.ones(100), input_range]).T
y = X @ beta + np.random.normal(0, 0.3, 100)
y[80] = 10

def calculate_laplacian_loss(parameters, X, y):
    return np.sum(np.abs(y - X @ parameters))

# Complete as linhas abaixo
beta_hat_gaussian = linalg.inv(X.T @ X) @ X.T @ y
beta_hat_laplacian = minimize(calculate_laplacian_loss, beta, args=(X, y))
```

```python
plt.scatter(X[:, 1], y, label='Dados Originais')
plt.plot(X[:, 1], X @ beta_hat_gaussian, label='Gaussiana', color='red')
plt.plot(X[:, 1], X @ beta_hat_laplacian.x, label='Laplaciana', color='green')
plt.xlabel('X')
plt.ylabel('Y')
plt.legend()
plt.show()
```

```python
# Calculando o erro:
error_gaussian = np.sqrt(np.sum((beta - beta_hat_gaussian)**2))
error_laplacian = np.sqrt(np.sum((beta - beta_hat_laplacian.x)**2))
```
